```python
positions = df['Position'].unique()

positions
```

```python
front_court = ['Centre',
       'Forward', 'Post Player', 'Post Position', 'Power Forward',
       'Center', 'Post /Center',
       'Power Forward Center', 'Power Forward ',
       'PowerForward']
```

```python
front_court_players = df.loc[df['Position'].isin(front_court)]

front_court_players.head()
```

```python
front_court_players['3 PA %'].replace({'#DIV/0!': 0}, inplace=True)
front_court_players['2 PA %'].replace({'#DIV/0!': 0}, inplace=True)
front_court_players['FREE THROW ACCURACY %'].replace({'#DIV/0!': 0}, inplace=True)
```

```python
attempts_stats = front_court_players['3 POINTS ATTEMPT'].describe()

attempts_stats
```
