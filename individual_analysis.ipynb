```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('player_stats_men.csv')
```

```python
df.columns
```

```python
df.sort_values(ascending=False, by=['TOTAL ASSIST']).head(20)
```

```python
top_scorers = df.sort_values(ascending=False, by=['TOTAL SCORE']).head(10)
top_scorers
```

```python
for _, player in top_scorers.iterrows():
    total_points = float(player['TOTAL SCORE'])
    three_percentage = float(player['3 POINTS MADE']) * 3 / total_points
    two_percentage = float(player['2 POINT MADE']) * 2 / total_points
    ft_percentage = float(player['FREE THROW MADE']) / total_points
    
    # Pie chart, where the slices will be ordered and plotted counter-clockwise:
    labels = '3s ({}%)'.format("{:.2f}".format(float(player['3 PA %']))), '2s ({}%)'.format("{:.2f}".format(float(player['2 PA %']))), 'FTs ({}%)'.format("{:.2f}".format(float(player['FREE THROW ACCURACY %'])))
    sizes = [three_percentage, two_percentage, ft_percentage]
    explode = (0.1, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax1.set_title(player['PLAYER NAME'] + ' ({})'.format(player['CLUB']) + ' TOTAL POINTS: {}'.format(player['TOTAL SCORE']))
    plt.savefig('./player_scoring_pie/{}.jpg'.format(player['PLAYER NAME']))
    plt.show()
```
